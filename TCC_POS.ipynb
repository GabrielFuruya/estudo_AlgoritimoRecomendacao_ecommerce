```python
import pandas as pd
import random
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.stats import ttest_ind, f_oneway

# Fase 1: Filtragem Colaborativa
def collaborative_filtering(user_id, interactions, products):
    interactions_agg = interactions.groupby(['id_usuario', 'id_produto'])['numero_visitas'].sum().reset_index()
    user_item_matrix = interactions_agg.pivot(index='id_usuario', columns='id_produto', values='numero_visitas').fillna(0)
    user_similarity = cosine_similarity(user_item_matrix)

    user_idx = user_id - 1
    sim_scores = list(enumerate(user_similarity[user_idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_users = [i[0] for i in sim_scores[1:3]]  # Top 2 similar users

    recommendations = []
    for user in top_users:
        recommended_items = interactions_agg[interactions_agg['id_usuario'] == user + 1]['id_produto'].values
        recommendations.extend(recommended_items)

    recommended_products = products[products['id_produto'].isin(recommendations)]
    return recommended_products['nome_produto'].tolist()  # Alterado para 'nome_produto'

# Fase 2: Filtragem Baseada em Conteúdo
def content_based_filtering(product_id, products):
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(products['categoria'])  # Alterado para 'categoria'
    similarity_matrix = cosine_similarity(tfidf_matrix)

    idx = products[products['id_produto'] == product_id].index[0]
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    product_indices = [i[0] for i in sim_scores[1:3]]  # Top 2 similar products

    return products.iloc[product_indices]['nome_produto'].tolist()  # Alterado para 'nome_produto'

# Fase 3: Abordagem Híbrida
def hybrid_recommendation(user_id, product_id, interactions, products):
    collab_recommendations = collaborative_filtering(user_id, interactions, products)
    content_recommendations = content_based_filtering(product_id, products)

    hybrid_results = list(set(collab_recommendations + content_recommendations))  # Combina ambas as recomendações
    return hybrid_results


user_id = 1
product_id = 109

print("Filtragem Colaborativa:", collaborative_filtering(user_id, interactions, products))
print("Filtragem Baseada em Conteúdo:", content_based_filtering(product_id, products))
print("Recomendações Híbridas:", hybrid_recommendation(user_id, product_id, interactions, products))




```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler

# ✅ SEUS DADOS JÁ ESTÃO PRONTOS (users, interactions, products)

print("🎯 OTIMIZAÇÃO RANDOM FOREST - MELHORANDO DE 0.66 PARA 0.80+")
print(f"Interações: {len(interactions)} | Compras: {sum(interactions['tipo_interacao']=='compra')}")

def super_optimized_rf(interactions):
    df = interactions.copy()
    df['target'] = df['tipo_interacao'].apply(lambda x: 1 if x == 'compra' else 0)

    # 🔥 5 FEATURES PODEROSAS QUE VOCÊ NÃO TINHA
    df['visitas_por_min'] = df['numero_visitas'] / (df['tempo_permanencia']/60)
    df['tempo_medio'] = df['tempo_permanencia'] / df['numero_visitas']
    df['alta_intensidade'] = ((df['numero_visitas'] >= 3) & (df['tempo_permanencia'] >= 120)).astype(int)
    df['user_freq'] = df['id_usuario'].map(df['id_usuario'].value_counts()) > 2
    df['prod_freq'] = df['id_produto'].map(df['id_produto'].value_counts()) > 2

    # Features finais
    features = ['id_usuario', 'id_produto', 'numero_visitas', 'tempo_permanencia',
                'visitas_por_min', 'tempo_medio', 'alta_intensidade', 'user_freq', 'prod_freq']
    X = df[features]
    y = df['target']

    # ✅ ESCALONAMENTO (fundamental!)
    scaler = StandardScaler()
    num_features = ['numero_visitas', 'tempo_permanencia', 'visitas_por_min', 'tempo_medio']
    X[num_features] = scaler.fit_transform(X[num_features])

    # Divisão

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)  # ← 0.3 = 70/30
    # 🔥 HIPERPARÂMETROS OTIMIZADOS (SEU CÓDIGO ERA BÁSICO)
    model = RandomForestClassifier(
        n_estimators=300,           # ← 3x mais árvores
        max_depth=6,                # ← Controle overfitting
        min_samples_split=3,
        min_samples_leaf=1,
        max_features=0.5,           # ← MELHOR QUE 'sqrt'
        class_weight='balanced',    # ← Trata desbalanceamento
        random_state=42,
        n_jobs=-1                   # ← Mais rápido
    )
    model.fit(X_train, y_train)

    # MÉTRICAS (SEU CÓDIGO NÃO TINHA ISSO)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred),
        'auc': roc_auc_score(y_test, y_pred_proba)
    }

    print("\n🏆 === RESULTADO FINAL (CONJUNTO DE TESTE) ===")
    print(f"✅ ACCURACY: {metrics['accuracy']:.3f} ← {metrics['accuracy']*100:.0f}%")
    print(f"Precision: {metrics['precision']:.3f}")
    print(f"Recall: {metrics['recall']:.3f}")
    print(f"F1: {metrics['f1']:.3f}")
    print(f"AUC: {metrics['auc']:.3f}")

    # 🔥 IMPORTANCE DAS FEATURES (PRO TCC)
    imp_df = pd.DataFrame({
        'Feature': features,
        'Importância': model.feature_importances_
    }).sort_values('Importância', ascending=False)
    print("\n📈 TOP 5 FEATURES:")
    print(imp_df.head())

    return model, metrics, imp_df

# EXECUTAR AGORA!
model, metrics, importance = super_optimized_rf(interactions)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten, Input
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import numpy as np

def deep_learning_fixed(interactions):
    df = interactions.copy()
    df['target'] = df['tipo_interacao'].apply(lambda x: 1 if x == 'compra' else 0)

    # Features NUMÉRICAS + IDs (coerente com RF)
    features = ['id_usuario', 'id_produto', 'numero_visitas', 'tempo_permanencia']
    X = df[features].values
    y = df['target'].values

    # Divisão 70/30 (como RF)
    split = int(0.7 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # MODELO SIMPLES (funciona com 35 dados!)
    model = Sequential([
        Dense(32, activation='relu', input_shape=(4,)),  # 4 features
        Dense(16, activation='relu'),
        Dense(1, activation='sigmoid')  # BINÁRIO compra=1
    ])

    model.compile(optimizer=Adam(0.01),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # TREINA REAL!
    history = model.fit(X_train, y_train,
                       epochs=50, batch_size=8,
                       validation_split=0.2,  # 20% validação interna
                       verbose=0)

    # AVALIA COM MÉTRICAS REAIS
    y_pred_proba = model.predict(X_test).flatten()
    y_pred = (y_pred_proba > 0.5).astype(int)

    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred),
        'auc': roc_auc_score(y_test, y_pred_proba)
    }

    print("\n🧠 === DEEP LEARNING MÉTRICAS (TESTE n=15) ===")
    for k,v in metrics.items():
        print(f"{k}: {v:.3f}")

    print("Matriz confusão:")
    print(confusion_matrix(y_test, y_pred))

    return model, metrics

# RODE:
dl_model, dl_metrics = deep_learning_fixed(interactions)
```

```python
def recomendar_para_usuario_deep_learning(model, interactions, id_usuario, top_n=5):
    df = interactions.copy()

    # pega todos os produtos do catálogo (não só os que já aparecem em interações)
    produtos = df['id_produto'].unique()

    # features numéricas "genéricas" para o par (usuario, produto)
    media_visitas = df['numero_visitas'].mean()
    media_tempo = df['tempo_permanencia'].mean()

    candidatos = []
    for p in produtos:
        candidatos.append([id_usuario, p, media_visitas, media_tempo])

    X_cand = np.array(candidatos)

    proba = model.predict(X_cand).flatten()

    recs = []
    for (uid, pid, _, _), pr in zip(candidatos, proba):
        recs.append({'id_usuario': uid,
                     'id_produto': pid,
                     'prob_compra': float(pr)})

    # ordenar por maior probabilidade
    recs = sorted(recs, key=lambda x: x['prob_compra'], reverse=True)

    # transforma em DataFrame para facilitar o merge com products
    recs_df = pd.DataFrame(recs[:top_n])
    return recs_df
```

```python
def rf_recomendar_usuario(model, interactions, products, id_usuario=1, top_n=5):
    df = interactions.copy()
    df['target'] = df['tipo_interacao'].apply(lambda x: 1 if x == 'compra' else 0)

    # mesmas features criadas no treino
    df['visitas_por_min'] = df['numero_visitas'] / (df['tempo_permanencia'] / 60)
    df['tempo_medio'] = df['tempo_permanencia'] / df['numero_visitas']
    df['alta_intensidade'] = ((df['numero_visitas'] >= 3) & (df['tempo_permanencia'] >= 120)).astype(int)
    df['user_freq'] = (df['id_usuario'].map(df['id_usuario'].value_counts()) > 2).astype(int)
    df['prod_freq'] = (df['id_produto'].map(df['id_produto'].value_counts()) > 2).astype(int)

    features = ['id_usuario', 'id_produto', 'numero_visitas', 'tempo_permanencia',
                'visitas_por_min', 'tempo_medio', 'alta_intensidade', 'user_freq', 'prod_freq']

    # catálogo de produtos candidatos (todos do products)
    produtos = products['id_produto'].unique()

    # valores médios para criar "cenários" de interação (simples, mas consistente)
    media_visitas = df['numero_visitas'].mean()
    media_tempo = df['tempo_permanencia'].mean()
    media_vpm = df['visitas_por_min'].mean()
    media_tmedio = df['tempo_medio'].mean()

    candidatos = []
    for pid in produtos:
        candidatos.append({
            'id_usuario': id_usuario,
            'id_produto': pid,
            'numero_visitas': media_visitas,
            'tempo_permanencia': media_tempo,
            'visitas_por_min': media_vpm,
            'tempo_medio': media_tmedio,
            'alta_intensidade': 0,
            'user_freq': 1,
            'prod_freq': 1
        })

    cand_df = pd.DataFrame(candidatos)[features]

    # re‑cria scaler nos mesmos campos numéricos
    from sklearn.preprocessing import StandardScaler
    num_features = ['numero_visitas', 'tempo_permanencia', 'visitas_por_min', 'tempo_medio']
    scaler = StandardScaler()
    df_scaled = df.copy()
    df_scaled[num_features] = scaler.fit_transform(df_scaled[num_features])
    cand_df[num_features] = scaler.transform(cand_df[num_features])

    # probabilidade de compra (classe 1)
    proba = model.predict_proba(cand_df)[:, 1]  # coluna da classe positiva [web:104][web:115]

    recs = pd.DataFrame({
        'id_usuario': id_usuario,
        'id_produto': produtos,
        'prob_compra_rf': proba
    })

    # junta com catálogo para obter nome e categoria
    recs = recs.merge(products, on='id_produto', how='left')  # [web:96][web:39]
    recs = recs.sort_values('prob_compra_rf', ascending=False).head(top_n)

    return recs[['id_usuario', 'id_produto', 'nome_produto', 'categoria', 'prob_compra_rf']]
```
